## description of the dataset
#### .step: represents a unit of time where 1 step equals 1 hour
#### .type: type of online transaction
#### .amount: the amount of the transaction
#### .nameOrig: customer starting the transaction
#### .oldbalanceOrg: balance before the transaction
#### .newbalanceOrig: balance after the transaction
#### .nameDest: recipient of the transaction
#### .oldbalanceDest: initial balance of recipient before the transaction
#### .newbalanceDest: the new balance of recipient after the transaction
#### .isFraud: fraud transaction

## Build the Model

### preprocessing 

```python
df['type']=df['type'].map({'PAYMENT':0,'CASH_IN':1,'DEBIT':2,'CASH_OUT':3,'TRANSFER':4})
```

```python
df.head()
```

### import libaries

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,roc_curve,auc,ConfusionMatrixDisplay
```

### put seed  and copy the dataset 

```python
seed=42
np.random.seed(seed)
random.seed(seed)

X=df.copy()
```

### drop irrlevent columns

```python
X.drop(['nameOrig','newbalanceOrig','nameDest','newbalanceDest','quantity','oldbalanceOrg_amt','oldbalanceDest_amt'],axis=1,inplace=True)
y=X.pop('isFraud')
```

```python
X
```

### Stratifed train-test spilt

```python
skfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=seed)

for train_idx,test_idx in skfold.split(X,y):
    X_train,X_test=X.iloc[train_idx],X.iloc[test_idx]
    y_train,y_test=y.iloc[train_idx],y.iloc[test_idx]
```

### Scaling the dataset

```python
Scaler=StandardScaler()
scaled_train=Scaler.fit_transform(X_train)
scaled_test=Scaler.fit_transform(X_test)

X_train=pd.DataFrame(scaled_train,index=X_train.index,columns=X_train.columns)
X_test=pd.DataFrame(scaled_test,index=X_test.index,columns=X_test.columns)
```

### UnderSampling the dataset 

```python
X_train,y_train=RandomUnderSampler(sampling_strategy='majority').fit_resample(X_train,y_train)
```

### Plot the countplot after under smapling

```python
y_res = pd.DataFrame(y_train, columns=['isFraud'])
ax = sns.countplot(x='isFraud', data=y_res, palette='rocket')
for container in ax.containers:
    ax.bar_label(container)

plt.title('Count of Fraud Transactions After Undersampling')
plt.xlabel('Is Fraud')
plt.ylabel('Number of Transactions')
plt.show()
```

### Comparison between Models Evaluation

```python
def model_comparison_evaluate(classifiers,X,y):
    print('K-Fold Cross-Validation:\n')
    for name,model in classifiers.items():
        print(f"{name}")
        scoring=['accuracy','precision','recall','f1','roc_auc']

        for score in scoring:
            scores=cross_val_score(model,X,y,scoring=score,cv=skfold,n_jobs=-1)
            print(f'Mean {score} score : {scores.mean():.3f} ({scores.std():.3f})')

            print('\n')
```

```python
classifiers={
    'Random Forest Classifier':RandomForestClassifier(class_weight='balanced',random_state=seed),
    'Logistic Regression':LogisticRegression(class_weight='balanced',random_state=seed),
    'DecisionTreeClassifier':DecisionTreeClassifier(class_weight='balanced',random_state=seed),
    'Support Vector Classifier':SVC(class_weight='balanced',random_state=seed)
}
```

```python
model_comparison_evaluate(classifiers,X_train,y_train)
```

### Random Forest Classifier

```python
model=RandomForestClassifier(class_weight='balanced',random_state=seed)
model.fit(X_train,y_train)

y_pred=model.predict(X_test)
y_pred_score=model.predict_proba(X_test)[:,1]
print("Random Forest Classifier :")
print(classification_report(y_pred,y_test,labels=[0,1],target_names=['Non-Fraud [0]','Fraud [1]']),'\n')
```

### plot the Confuision Matrix

```python
fig,ax=plt.subplots(1,2,figsize=(20,5))
ax[0].set_title('Confusion Matrix Of Random Forest Model :')
ConfusionMatrixDisplay.from_predictions(y_test,y_pred,values_format='',ax=ax[0],cmap='rocket')
ax[0].grid(visible=False)

FPr,TPr,thresholds=roc_curve(y_test,y_pred_score)
roc_auc=auc(FPr,TPr)
ax[1].set_title('ROC Curve - Random Forest Classifier')
ax[1].plot(FPr,TPr,label='AUC = %0.3f' % roc_auc)
ax[1].plot([0,1],[0,1],'--',c='lightsteelblue')
ax[1].legend(loc='lower right')
ax[1].set_ylabel('True Positive Rate')
ax[1].set_xlabel('False Positive Rate')

plt.show()
```
