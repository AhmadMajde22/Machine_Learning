### Data Preprocessing

```python
def object_to_int(data):
    if data.dtype =='object':
        data=LabelEncoder().fit_transform(data)
    return data
```

```python
df=df.apply(lambda x : object_to_int(x))
df.head()
```

```python
df.corr()['Churn'].sort_values(ascending=False)
```

```python
X=df.drop(columns=['Churn'])
y=df['Churn']
```

```python
X
```

```python
y 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=41,stratify=y)
```

```python
def distplot(feature,frame,color='r'):
    plt.figure(figsize=(8,3))
    plt.title(f"Dist for {feature}")
    ax=sns.distplot(frame[feature],color=color)
```

```python
cols=['tenure','MonthlyCharges','TotalCharges']

for n in cols:
    distplot(n,df)
```

```python
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
scaler=pd.DataFrame(StandardScaler().fit_transform(df[cols].astype('float64')),columns=cols)

for col in numerical_cols:
    distplot(col,scaler,color='c')
```

```python
cate_cols_ohot=['PaymentMethod','Contract','InternetService']
cate_cols_LE=list(set(X_train.columns)- set(cols) - set(cate_cols_ohot))
```

```python
cate_cols_ohot
```

```python
cate_cols_LE
```

```python
Scaler=StandardScaler()

X_train[cols] = Scaler.fit_transform(X_train[cols])
X_test[cols]=Scaler.transform(X_test[cols])
```

## Machine Learning Models

### KNN model

```python
knn_model=KNeighborsClassifier(n_neighbors=11)
knn_model.fit(X_train,y_train)
predicted_y=knn_model.predict(X_test)
accuracy_knn=knn_model.score(X_test,y_test)
print("KNN accuracy : ",accuracy_knn)
```

```python
print(classification_report(y_test,predicted_y))
```

### SVC model

```python
svc=SVC(random_state=42)
svc.fit(X_train,y_train)
predicted_y=svc.predict(X_test)
accuracy_svc=svc.score(X_test,y_test)
print("SVM accuracy is :",accuracy_svc)
```

```python
print(classification_report(y_test,predicted_y))
```

### Random Forest 

```python
RF_model=RandomForestClassifier(n_estimators=500,oob_score=True,n_jobs=-1,random_state=42,max_features='log2',max_leaf_nodes=300,max_depth=100)
RF_model.fit(X_train,y_train)
predicted_y=RF_model.predict(X_test)
accuracy_RF=RF_model.score(X_test,y_test)
print("Random Forest accuracy is :",accuracy_RF)
```

```python
print(classification_report(y_test,predicted_y))
```

```python
plt.figure(figsize=(4,3))

sns.heatmap(confusion_matrix(y_test,predicted_y),annot=True,fmt='d',cmap='coolwarm')

plt.title("Random Forest Heatmap")

plt.show()
```

```python
y_pred_prob=RF_model.predict_proba(X_test)[:,1]
fpr_rf,tpr_rf,threshold=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rf,tpr_rf,label='Random Forest',color='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Random Forest ROC Curve",fontsize=14)
plt.show()
```

### Logistic Regression 

```python
lr_model=LogisticRegression(penalty='l2',solver='sag')
lr_model.fit(X_train,y_train)
accuracy_lr=lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

```python
lr_pred=lr_model.predict(X_test)
print(classification_report(y_test,lr_pred))
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test,lr_pred),annot=True,fmt='d')
```

```python
y_pred_prob=lr_model.predict_proba(X_test)[:,1]
fpr,tpr,threshold=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'b--')
plt.plot(fpr,tpr,label='Logistic Regression',color='r')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Logistic Regression ROC Curve")

plt.show()
```

### Decision Tree Classifier 

```python
tree_model=DecisionTreeClassifier(min_samples_leaf=100,min_samples_split=20)
tree_model.fit(X_train,y_train)
predicted_y2=tree_model.predict(X_test)
accuracy_tree=tree_model.score(X_test,y_test)
print("Decision Tree accuracy is",accuracy_tree)
```

```python
print(classification_report(y_test,predicted_y2))
```

### Voting classifier

```python
from sklearn.ensemble import VotingClassifier
Cla1=GradientBoostingClassifier()
Cla2=LogisticRegression()
#Cla3=DecisionTreeClassifier()
Cla4=AdaBoostClassifier()
#Cla5=RandomForestClassifier()
Classifier=VotingClassifier(estimators=[('GBC',Cla1),('LR',Cla2),('ABC',Cla4)],voting='soft')


Classifier.fit(X_train,y_train)
```

```python
prediction=Classifier.predict(X_test)
print("Voting Accuracy Score is ",accuracy_score(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

```python
plt.figure(figsize=(4,3))

sns.heatmap(confusion_matrix(y_test,prediction),annot=True,fmt='d')

plt.title("Voting Confusion Matrix")
plt.show()
```

```python
y_pred_prob=Classifier.predict_proba(X_test)[:,1]
fpr,tpr,threshold=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'b--')
plt.plot(fpr,tpr,label='GradientBoosting Classifier',color='r')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("GradientBoosting Classifier ROC Curve")

plt.show()
```
